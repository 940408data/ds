```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
s=pd.Series(np.arange(100))
s.plot(color="r",linestyle='--',label='abc',alpha=0.7,xlim=[1,150],xticks=[40,50,60],logy=True)
```
