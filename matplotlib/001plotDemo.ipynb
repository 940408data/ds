```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style="darkgrid")  ## 在前，中文设置在后
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
def plot_simple():
    #数据
    x=np.linspace(0,20,1000)
    sin_y=np.sin(x)
    cos_y=np.cos(x)
    
    #画图
    plt.plot(x,sin_y,"r--",label="siny")
    plt.plot(x,cos_y,"g",label="cosy")
    
    plt.xlabel("x")
    plt.xlim(0,20)
    
    plt.xlabel("y")
#     plt.xlim(0,20)
    plt.legend(loc="best")    
```

```python
plot_simple()
```
