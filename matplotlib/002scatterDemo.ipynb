```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style="darkgrid")  ## 在前，中文设置在后
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 中文 
mpl.rcParams['axes.unicode_minus'] = False  # 负坐标轴
%matplotlib inline
```

### 两标签的scatter

直线分割

```python
df=pd.DataFrame()
df["num1"]=np.random.normal(0,10,size=100)
df["num2"]=np.random.normal(0,10,size=100)
df["label"]=np.where(df.num1+df.num2>10,1,0)
x=np.linspace(-30,30,100)
y=-x+10
```

```python
plt.scatter(df[df.label==1].num1,df[df.label==1].num2)
plt.scatter(df[df.label==0].num1,df[df.label==0].num2)
plt.plot(x,y)
plt.xlabel("语文")
plt.ylabel("数学")
# plt.scatter(df.num1,df.num2)
```

抛物线分割

```python
df1=pd.DataFrame()
df1["num1"]=np.random.randint(0,30,100)
df1["num2"]=np.random.randint(0,300,100)
df1["label"]=np.where(df1.num2>(df1.num1-15)**2,1,0)
x=np.linspace(0,30,100)
y=(x-15)**2+10
```

```python
plt.scatter(df1[df1.label==0].num1,df1[df1.label==0].num2)
plt.scatter(df1[df1.label==1].num1,df1[df1.label==1].num2)
plt.plot(x,y)
plt.show()
```
